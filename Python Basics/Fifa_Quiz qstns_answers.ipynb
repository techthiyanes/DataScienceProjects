# Q2) What is the arithmetic mean of the variable ‘Value’ (in Euros)?

```python
df['Value'].head()
```

```python
df['Unit']=df['Value'].str[-1]
df['Unit'].head()
```

# Q11: What is the average wage one can expect to pay for the top 5 players in the following positions namely RCB, LCM,CB,LDM,CAM?

```python
df['Wage'].head()
```

```python
wage=df.Wage.str.split("€",n=-2,expand=True)[1]
wage.head()
```

```python
wage=wage.str.replace('K','000')
wage.head()
```

```python
df.Wage.str.split("€",n=-2,expand=True)[0].head()
```

```python
wage=list(np.float_(wage))
df.Wage=wage
df.Wage.head()
```

```python
topwages = df[['Position','Wage','Overall']].sort_values(by='Overall',ascending=False).groupby('Position').head(5)
topwages
```

```python
selectedwages=topwages[['Position','Wage']].groupby('Position').mean()
```

```python
selectedwages.head()
```
