## Q25. Make the following ranges of miles ["<5","5-10","10-20","20-35",">35"]. Keep these in the horizontal axis and count of trips in the vertical axis. Plot a bar plot. (2 points)

```python
auto=pd.read_csv("C:/Users/Thiyaneswaran_Manoha/Downloads/uberdrive.csv")
# print(auto['MILES*'].arange(0,5).value_counts())
#https://stackoverflow.com/questions/25010215/pandas-groupby-how-to-compute-counts-in-ranges
ranges = [0,5,10,20,35,200]
count=auto['MILES*'].groupby(pd.cut(auto['MILES*'], ranges)).count()
print(count)
count.to_csv('C:/Users/Thiyaneswaran_Manoha/Downloads/rangecount.csv')
rangecsv=pd.read_csv('C:/Users/Thiyaneswaran_Manoha/Downloads/rangecount.csv')
rangecsv.columns = ['Range of miles','Count of trips']
print(rangecsv)

sns.barplot(x = "Range of miles", y = "Count of trips", hue = "Range of miles", data = rangecsv)
plt.xlabel("Range of miles")
plt.ylabel("COUNT OF DAY OF TRIP")
plt.title("Range of miles vs Count of trips")
plt.show()


new_auto = auto.dropna(axis = 0, how ='any') 
print("new_data size***",new_auto.size)
count=new_auto['MILES*'].groupby(pd.cut(new_auto['MILES*'], ranges)).count()
print(count)
count.to_csv('C:/Users/Thiyaneswaran_Manoha/Downloads/rangecount1.csv')
rangecsv1=pd.read_csv('C:/Users/Thiyaneswaran_Manoha/Downloads/rangecount1.csv')
rangecsv1.columns = ['Range of miles1','Count of trips1']
print(rangecsv1)

sns.barplot(x = "Range of miles1", y = "Count of trips1", hue = "Range of miles1", data = rangecsv1)
plt.xlabel("Range of miles1")
plt.ylabel("COUNT OF DAY OF TRIP")
plt.title("Range of miles vs Count of trips")
plt.show()
```
